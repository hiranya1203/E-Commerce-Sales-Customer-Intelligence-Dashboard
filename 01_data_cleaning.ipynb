```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("online_retail_II.csv")
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
# remove cancelled orders
df = df[~df['Invoice'].str.startswith('C')]

```

```python
print("Cancelled orders:", df['Invoice'].str.startswith('C').sum())
print("Total rows:", len(df))
```

Cancelled transactions(identified by Invoice numbers starting with 'C') were removed, reducing that dataset from 1,067,371 to 1,047,877 rows, ensuring accurate sales and customer analysis.

```python
# drop null values in Customer ID
df = df.dropna(subset=['Customer ID'])

print("Remaining nulls:", df['Customer ID'].isnull().sum())
print("Total rows:", len(df))
```

Customer ID is essential for analysing customer behavior. Rows without Customer ID cannot be linked to any customer, so they were removed to ensure accurate customer-level insights.

```python
# remove rows where Quantity and Price <= 0
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)]

print("Invalid Quantity rows:", (df['Quantity'] <= 0).sum())
print("Invalid Price rows:", (df['Price'] <= 0).sum())
print("Total rows:", len(df))
```

Removed rows where Quantity and Price <= 0, to eliminate returns and erroneous transactions, ensuring accyrate revenue analysis.

```python
print(df['Quantity'].min())
print(df['Price'].min())
```

```python
print("Total rows:", len(df))
```

```python
# Fix datatypes
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df['Customer ID'] = df['Customer ID'].astype(int).astype(str)
```

```python
df.dtypes
```

```python
# Create revenue column
df['Revenue'] = df['Quantity'] * df['Price']
```

```python
df[['Quantity', 'Price', 'Revenue']].head()
```

```python
# Extract time features
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Month_Name'] = df['InvoiceDate'].dt.strftime('%b')
df['Week'] = df['InvoiceDate'].dt.isocalendar().week
df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()
```

```python
df[['InvoiceDate', 'Year', 'Month', 'Month_Name', 'Week', 'DayOfWeek']].head()
```

```python
df.to_csv('retail_clean.csv', index=False)
print("File Saved")
```
